# Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_style('whitegrid')
```

## Read Data

```python
df = pd.read_csv('../processed/processed_customers_raw.csv')
df.head()
```

# Exploratory Data Analysis

## Count of each categorical Variable

```python
print("\nEducation Distribution:")
print(df['Education'].value_counts())
```

```python
print("\nMarital Status Distribution:")
print(df['Marital_Status'].value_counts())
```

```python
print("\nKidhome vs Teenhome (Family Info):")
print(df[['Kidhome', 'Teenhome']].value_counts().sort_index())
```

## Visualization of Correlation Matrix

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f', vmin = -1, vmax = 1)
plt.title("Correlation Heatmap")
plt.show()
```

It seems *Z_Cost_Contact* and *Z_Revenue* is not affected by or affects other attributes.

## Visualization of *Education*

```python
sns.countplot(data = df, x = 'Education')
plt.title('Countplot of Education')
plt.show()
```

The dataset consists of more graduated individuals.

## Plot of Average Income by Eduction

```python
df.groupby('Education')['Income'].mean().plot()
plt.title('Average Income by Education')
plt.xlabel('Education')
plt.ylabel('Average Income')
plt.show()
```

People with Basic Education seems to have lowest income. The better the education level, higher is the income.

## Visualization of *Marital Status*

```python
sns.countplot(data = df, x = 'Marital_Status')
plt.title('Marital Status Count')
plt.show()
```

## Count of Responses given by Customers

```python
sns.countplot(data = df, x = 'Response')
plt.title('Count of Response in the data')
plt.show()
```

Only a few customers seem to respond.
